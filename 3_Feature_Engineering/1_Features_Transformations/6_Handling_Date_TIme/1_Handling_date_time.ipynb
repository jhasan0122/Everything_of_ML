```python
import numpy as np
import pandas as pd
```

# Working with Date

```python
dateData = pd.read_csv('orders.csv')
dateData
```

```python
dateData.info()
```

```python
dateData['date'] = pd.to_datetime(dateData['date'])
dateData.info()
```

# 1. Extract year

```python
dateData['year'] = dateData['date'].dt.year
dateData
```

# 2. Extract month

```python
dateData['month'] = dateData['date'].dt.month
dateData
```

# 3.Extract Monthname

```python
dateData['month_name'] = dateData['date'].dt.month_name()
dateData
```

# 4. Extract date

```python
dateData['date_day'] = dateData['date'].dt.day
dateData
```

# 5. Day of week

```python
dateData['day_week'] = dateData['date'].dt.dayofweek
dateData
```

# 6. Day name

```python
dateData['day_name'] = dateData['date'].dt.day_name()
dateData
```

# 7. IS weekend

```python
dateData['is_weekend'] = np.where(dateData['day_name'].isin(['Saturday','Sunday']),1,0)
dateData
```

# 8.Extract week of the year

```python
dateData['date_week'] = dateData['date'].dt.isocalendar().week
dateData
```

```python
dateData['quarter'] = dateData['date'].dt.quarter
dateData
```

# 9.Extract semester

```python
dateData['semester'] = np.where(dateData['quarter'].isin([1,2]),1,2)
```

# 10. Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
today - dateData['date']
```

```python
(today - dateData['date']).dt.days
```

# Working with time

```python
timeData = pd.read_csv('messages.csv')
timeData
```

```python
timeData.info()
```

```python
timeData['date'] = pd.to_datetime(timeData['date'])
timeData.info()
```

# 1. Extract hour, min and seconds

```python
timeData['hour'] = timeData['date'].dt.hour
timeData['min'] = timeData['date'].dt.minute
timeData['sec'] = timeData['date'].dt.second
timeData
```

# 2. Extract time part

```python
timeData['time'] = timeData['date'].dt.time
timeData
```

# 3. Time difference

```python
today - timeData['date']
```

# 4. Time difference in sec

```python
(today - timeData['date'])/np.timedelta64(1,'s')
```

# 5. Time difference in min

```python
(today - timeData['date'])/np.timedelta64(1,'m')
```

# 6.TIme diff in hour

```python
(today - timeData['date'])/np.timedelta64(1,'h')
```
