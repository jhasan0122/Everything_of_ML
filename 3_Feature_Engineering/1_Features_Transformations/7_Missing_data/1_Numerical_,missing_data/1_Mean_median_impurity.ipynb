```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_toy.csv')
df
```

```python
df.isnull().mean()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=0)
```

```python
X_train.isnull().mean()
```

# 1. Using Pandas

### 1. Check mean median

```python
mean_age = X_train['Age'].mean()
print(mean_age)
X_train["Age_mean"] = X_train['Age'].fillna(mean_age)
X_train
```

```python
median_age = X_train['Age'].median()
print(median_age)
X_train["Age_median"] = X_train['Age'].fillna(median_age)
X_train
```

```python
mean_fare = X_train['Fare'].mean()
print(mean_fare)
X_train["Fare_mean"] = X_train['Fare'].fillna(mean_fare)
X_train
```

```python
median_fare = X_train['Fare'].median()
print(median_fare)
X_train["Fare_median"] = X_train['Fare'].fillna(median_fare)
X_train
```

### 2. Check mean median

```python
X_train[['Age','Age_mean','Age_median','Fare','Fare_mean','Fare_median']].var()
```

### 3. Check distribution

```python
sns.kdeplot(data=X_train,x='Age')
sns.kdeplot(data=X_train,x='Age_mean')
sns.kdeplot(data=X_train,x='Age_median')
```

```python
sns.kdeplot(data=X_train,x='Fare')
sns.kdeplot(data=X_train,x='Fare_mean')
sns.kdeplot(data=X_train,x='Fare_median')
```
